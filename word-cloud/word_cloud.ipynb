```python
import wordcloud
import matplotlib.pyplot as plt

from imageio import imread

my_font = 'font/wqy-microhei.ttc'
```

```python
with open('../tf-idf/stage1.txt', 'r', encoding='utf-8') as f:
    file_data = ''
    for line in f:
        file_data += line
cloud_obj = wordcloud.WordCloud(font_path=my_font, 
                                background_color='white', mask=imread('image/01.png')).generate(file_data)
plt.imshow(cloud_obj)
plt.axis('off')
cloud_obj.to_file('image/01_res.png')
```
